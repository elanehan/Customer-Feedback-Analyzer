```python
# Cell 1: Setup and Initialization

import os
import json
from dotenv import load_dotenv
from IPython.display import Markdown, display

# --- Load Environment Variables ---
# This line loads the secrets from your `backend/.env` file.
# It's crucial that you run this notebook from the root of your project directory.
print("Loading environment variables from backend/.env...")
load_dotenv(dotenv_path='backend/.env')
print("Environment variables loaded.")

# --- Import the Agent ---
# This imports the final, compiled agent from the graph.py file you just created.
# If you make changes to your agent's logic, you may need to restart the notebook kernel
# or re-run this cell to ensure the latest version is imported.
try:
    from backend.src.agent.graph import agent_executor
    print("Agent executor imported successfully.")
except Exception as e:
    print(f"Error importing agent executor: {e}")
    print("Please ensure your backend code has no syntax errors and all dependencies are installed.")
```

```python
# Cell 2: Run the Full Analysis (Final, Robust Version)
from datetime import date, datetime
from pydantic import BaseModel

# This helper function now knows how to handle both datetimes AND Pydantic models
def json_serial(obj):
    """JSON serializer for objects not serializable by default json code"""
    if isinstance(obj, (datetime, date)):
        return obj.isoformat()
    # This is the new part that handles Pydantic models
    if isinstance(obj, BaseModel):
        return obj.model_dump() # Convert the Pydantic model to a dictionary
    raise TypeError ("Type %s not serializable" % type(obj))

# --- 1. Define Your Input ---
test_inputs = {"product_id": "B007JFMH8M"} # <-- REPLACE WITH A REAL PRODUCT ID

# --- 2. Execute the Agent using .stream() ---
final_state = None
print(f"🚀 Kicking off agent for Product ID: {test_inputs['product_id']}...\n")

try:
    for chunk in agent_executor.stream(test_inputs, {"recursion_limit": 10}):
        node_name = list(chunk.keys())[0]
        node_output = chunk[node_name]
        
        print(f"--- ✅ Node '{node_name}' Finished ---")
        
        # This will now work correctly for all our data types
        print(json.dumps(node_output, indent=2, sort_keys=True, default=json_serial))
        print("\n" + "="*80 + "\n")
        
        final_state = chunk

    # --- 3. Display the Final Generated Summary ---
    if final_state and "generate_summary" in final_state:
        print("--- ✨ FINAL SUMMARY ✨ ---")
        summary = final_state['generate_summary'].get('summary', 'No summary was generated.')
        display(Markdown(summary))
    else:
        print("--- AGENT FAILED TO PRODUCE A FINAL SUMMARY ---")

except Exception as e:
    print(f"An error occurred during agent execution: {e}")
```
